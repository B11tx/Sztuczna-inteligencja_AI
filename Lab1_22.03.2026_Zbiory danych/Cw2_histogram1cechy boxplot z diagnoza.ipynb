```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df_cancer = pd.read_csv('data.csv')


sns.set_theme(style="whitegrid")

plt.figure(figsize=(14, 6))

print("--- BREAST CANCER DATASET ---")
print(f"Liczba próbek: {df_cancer.shape[0]}")
print(f"Liczba kolumn: {df_cancer.shape[1]}")
print("Nazwy kolumn:", df_cancer.columns.tolist())
print("\nRozkład klas M/B:")
print(df_cancer['diagnosis'].value_counts())



plt.subplot(1, 2, 1)
sns.histplot(data=df_cancer, x='radius_mean', kde=True, color='skyblue')
plt.title('Histogram: Rozkład radius_mean')


plt.subplot(1, 2, 2)

sns.boxplot(data=df_cancer, x='diagnosis', y='radius_mean', hue='diagnosis', palette='Set2', legend=False)
plt.title('Boxplot: Promień vs Diagnoza')
plt.xlabel('Diagnoza (M = Złośliwy, B = Łagodny)')
plt.ylabel('Średni promień')

plt.tight_layout()
plt.show()



```
