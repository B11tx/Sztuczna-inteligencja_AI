```python
from typing import Dict

def calculate_risk_score(data: Dict[str, float]) -> float:
    """
    Oblicza wynik ryzyka na podstawie słownika parametrów.
    
    Wymagane klucze: 'age', 'cholesterol', 'max_heart_rate'.
    """
    # Walidacja danych
    if any(val < 0 for val in data.values()) or data.get('max_heart_rate', 0) <= 0:
        raise ValueError('Wszystkie parametry muszą być dodatnie, a tętno > 0.')

    # Definicja wag 
    weights = {
        'age': 0.2,
        'cholesterol': 0.05,
        'max_heart_rate': -0.03
    }
    
    score = sum(data[key] * weights[key] for key in weights)
    return round(score, 2)


patient_data = {'age': 45, 'cholesterol': 220, 'max_heart_rate': 160}
print(f"Zrefaktoryzowany wynik: {calculate_risk_score(patient_data)}")
```
