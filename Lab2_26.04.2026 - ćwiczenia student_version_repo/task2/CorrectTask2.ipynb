```python
import unittest

def calculate_risk_score(age: int, cholesterol: int, max_heart_rate: int) -> float:
    if age < 0 or cholesterol < 0 or max_heart_rate <= 0:
        raise ValueError('All parameters must be positive.')

    score = age * 0.2 + cholesterol * 0.05 - max_heart_rate * 0.03
    return round(score, 2)

class TestRiskScore(unittest.TestCase):
    
    def test_standard_case(self):
        self.assertAlmostEqual(calculate_risk_score(40, 200, 70), 15.9, places=2)

    def test_zero_values(self):
        self.assertAlmostEqual(calculate_risk_score(0, 0, 60), -1.8, places=2)

    def test_negative_age_raises_error(self):
        with self.assertRaises(ValueError):
            calculate_risk_score(-5, 200, 70)
#Moj edge case: bardzo wysoki cholesterol
    def test_extreme_cholesterol(self):
        score = calculate_risk_score(50, 500, 150)
        self.assertIsInstance(score, float)
        self.assertGreater(score, 0)


if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
