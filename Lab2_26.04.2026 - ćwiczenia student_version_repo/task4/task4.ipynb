```python
import pandas as pd
import os

def load_data(file_name: str) -> pd.DataFrame:
    """Wczytuje dane z lokalnego pliku CSV z walidacją."""
    
    if not os.path.exists(file_name):
        raise FileNotFoundError(f"Błąd: Plik '{file_name}' nie został znaleziony w folderze projektu.")
    
    df = pd.read_csv(file_name)
    return df

def generate_expanded_report(df: pd.DataFrame) -> dict:
    """Rozbudowany raport statystyczny pod kątem AI."""
    report = {
        'liczba_pacjentow': len(df),
        'braki_w_danych': int(df.isna().sum().sum()),
        'sredni_wiek': round(df['age'].mean(), 1),
        'max_cholesterol': df['chol'].max(),
        
        'balans_klas': df['target'].value_counts(normalize=True).to_dict()
    }
    
   
    correlations = df.corr()['target'].sort_values(ascending=False)
    report['najsilniejsza_korelacja'] = correlations.index[1] 
    return report

if __name__ == '__main__':
  
    DATA_PATH = 'Data.csv'
    
    try:
        heart_df = load_data(DATA_PATH)
        print("--- ROZBUDOWANY RAPORT ANALIZY DANYCH ---")
        
        results = generate_expanded_report(heart_df)
        for key, value in results.items():
            print(f"{key}: {value}")
            
    except Exception as e:
        print(f"Wystąpił problem: {e}")
```
