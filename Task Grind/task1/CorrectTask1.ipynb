```python
def average_age(records: list) -> float:

    if not records:
        return 0.0

    total = 0
    for record in records:
        total += record.get('age', 0)

    return total / len(records)


if __name__ == '__main__':

    sample = [
        {'name': 'Anna', 'age': 21},
        {'name': 'Piotr', 'age': 23},
        {'name': 'Maria', 'age': 25},
    ]
    print(f"Średnia wieku (standard): {average_age(sample)}")


    empty_sample = []
    print(f"Średnia wieku (pusta lista): {average_age(empty_sample)}")
```
